```python
import pickle
import scipy as sp
import numpy as np
import pandas as pd
import anndata
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
# from samalg import SAM
# from samalg.gui import SAMGUI
# import samalg.utilities as ut

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))

from samalg import SAM
sam = SAM()
# total combo atlas
sam.load_data('data/brain_combined_samobj.h5ad')
genes = sam.adata.var_names

def target_to_ind(target,data,genes):
    '''helper function for finding contig index'''
    if type(target) is str:
        target=[target]
    target_ind = np.zeros(len(target))
    for i, contig in enumerate(genes):
        for j, targ in enumerate(target):
            if targ in contig:
                target_ind[j] = i
    return target_ind.astype(int)
```

# Expression Plots
* Define contigs you want to plot in the gene list
* Will run using the cells and projection in the active window of the GUI above

```python
genes=['dd_Smed_v4_659_0_1','_4389_0_1','_2072_0_1','_10835_0_1']

def custom_plot(gene,sam,ax=None,k_mean=True,sort=True,**kwargs):
    if ax is None:
        fig,ax = plt.subplots(ncols=1)
    dat = sam.adata.obsm["X_umap"]
    xx=dat[:,0]
    yy=dat[:,1]
    ind = target_to_ind(gene,sam,sam.adata.var_names)
    print(ind)
    if k_mean:
        c = np.squeeze(sam.adata.layers["X_knn_avg"][:,ind].toarray())
    else:
        c = np.squeeze(sam.adata_raw.X[:,ind].toarray())
    
    if sort:
        ind=np.argsort(c)
    else:
        ind = np.arange(c.size)
    ax.scatter(xx[ind],yy[ind],rasterized=True,c=c[ind],**kwargs,cmap='Reds')
    ax.set_title(gene)
    
fig,ax = plt.subplots(1,len(genes),dpi=300,sharex=True,sharey=True)
fig.set_size_inches((20,8))
for a,g in zip(ax,genes):
    custom_plot(g,sam_gui.current_sam,a,s=10)
# ax[0].set_xlim(-5,15)
# ax[0].set_ylim(-3,10)
plt.rcParams['svg.fonttype']='none'
fig.savefig('PLOTS/expression_UMAP.svg')
```
